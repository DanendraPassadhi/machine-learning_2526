## Soal 2 - Analisis Visual (15 poin)
1. Fenomena apa yang terjadi pada distribusi data 'age'?
2. Jika terdapat data yang hilang pada variabel 'age', strategi apa yang Anda terapkan? Mengapa?
3. Berapa jumlah outlier pada setiap kategori 'income' berkaitan dengan 'hour-per-week'? Kategori apa yang paling banyak memiliki outlier?

```python
# Jawab dengan komentar python

'''
  Bisa dengan multiple comment
  seperti ini
'''

# Jawaban Nomor 1
'''
    Distribusi data 'age' cenderung berbentuk positif skew, di mana sebagian besar data berada di sisi kiri, 
    ini menunjukkan bahwa data di dominasi oleh populasi berusia dewasa muda.
'''

# Jawaban Nomor 2
'''
    Jika terdapat data yang hilang, strategi yang diterapkan adalah mengisi nilai dengan mean. Namun dapat juga
    mengisi nilai dengan median ataupun modus. Karena, jika menggunakan mean, data diisi dari rata-rata nilai yang ada.
'''

# Jawaban Nomor 3
income_categories = df['income'].unique()

for category in income_categories:
    print(f"Analysis for Income Category: {category}")
    
    subset_df = df[df['income'] == category]
    
    Q1 = subset_df['hours-per-week'].quantile(0.25)
    Q3 = subset_df['hours-per-week'].quantile(0.75)
    IQR = Q3 - Q1
    
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR
    
    outliers = subset_df[
        (subset_df['hours-per-week'] < lower_fence) | 
        (subset_df['hours-per-week'] > upper_fence)
    ]
    
    print(f" Jumlah Outlier ditemukan: {len(outliers)}")


plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='income', y='hours-per-week')
plt.title('Hours-per-Week berdasarkan Income')
plt.xlabel('Income')
plt.ylabel('Hours per Week')
plt.show()


print('Outlier yang memiliki paling banyak otlier adalah individu dengan pendapatan <=50K dengan 7809 outlier.')
```
