# Praktikum 1 membuat JST sederhana (2 layer) dengan forward pass dan backpropagation manual. 

```python
import numpy as np

# Dataset XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Parameter
input_size = 2
hidden_size = 2
output_size = 1
lr = 0.1

# Inisialisasi bobot
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

# Fungsi aktivasi
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Training
for epoch in range(10000):
    # Forward pass
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    # Hitung error
    error = y - a2

    # Backpropagation
    d_a2 = error * sigmoid_derivative(a2)
    d_W2 = np.dot(a1.T, d_a2)
    d_b2 = np.sum(d_a2, axis=0, keepdims=True)

    d_a1 = np.dot(d_a2, W2.T) * sigmoid_derivative(a1)
    d_W1 = np.dot(X.T, d_a1)
    d_b1 = np.sum(d_a1, axis=0, keepdims=True)

    # Update bobot
    W1 += lr * d_W1
    b1 += lr * d_b1
    W2 += lr * d_W2
    b2 += lr * d_b2

    if epoch % 1000 == 0:
        loss = np.mean(np.square(error))
        print(f"Epoch {epoch}, Loss: {loss}")

# Output akhir
print("Prediksi:")
print(a2)
```

## Tugas 1:
- Ubah jumlah neuron hidden layer menjadi 3.
- Bandingkan hasil loss dengan konfigurasi awal.
- Tambahkan fungsi aktivasi ReLU dan bandingkan hasil.

```python
import numpy as np

# Dataset XOR
X_relu = np.array([[0,0],[0,1],[1,0],[1,1]])
y_relu = np.array([[0],[1],[1],[0]])

# Parameter
input_size_relu = 2
hidden_size_relu = 3
output_size_relu = 1
lr_relu = 0.1

# Inisialisasi bobot
W1_relu = np.random.randn(input_size_relu, hidden_size_relu)
b1_relu = np.zeros((1, hidden_size_relu))
W2_relu = np.random.randn(hidden_size_relu, output_size_relu)
b2_relu = np.zeros((1, output_size_relu))

# Fungsi aktivasi
def sigmoid_relu(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative_relu(x):
    return x * (1 - x)

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return (x > 0).astype(float)

# Training
for epoch in range(10000):
    # Forward pass
    z1_relu = np.dot(X_relu, W1_relu) + b1_relu
    a1_relu = relu(z1_relu) 
    z2_relu = np.dot(a1_relu, W2_relu) + b2_relu
    a2_relu = sigmoid_relu(z2_relu)

    # Hitung error
    error_relu = y_relu - a2_relu

    # Backpropagation
    d_a2_relu = error_relu * sigmoid_derivative_relu(a2_relu)
    d_W2_relu = np.dot(a1_relu.T, d_a2_relu)
    d_b2_relu = np.sum(d_a2_relu, axis=0, keepdims=True)

    d_a1_relu = np.dot(d_a2_relu, W2_relu.T) * relu_derivative(z1_relu)
    d_W1_relu = np.dot(X_relu.T, d_a1_relu)
    d_b1_relu = np.sum(d_a1_relu, axis=0, keepdims=True)

    # Update bobot
    W1_relu += lr_relu * d_W1_relu
    b1_relu += lr_relu * d_b1_relu
    W2_relu += lr_relu * d_W2_relu
    b2_relu += lr_relu * d_b2_relu

    if epoch % 1000 == 0:
        loss_relu = np.mean(np.square(error_relu))
        print(f"Epoch {epoch}, Loss: {loss_relu:.6f}")

# Output akhir
print("\nPrediksi:")
print(np.round(a2_relu, 4))
```
